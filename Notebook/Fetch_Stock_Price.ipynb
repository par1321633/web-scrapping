```python
time.sleep(3)
```

```python
def scrap_nyse_web(driver, url):
    time.sleep(5)
    dates_input_fields = driver.find_elements_by_class_name('react-datepicker__input-container')
    print (dates_input_fields)
    input_1 = dates_input_fields[0].find_element_by_tag_name('input')
    input_1.send_keys(Keys.CONTROL + "a")
    input_1.send_keys(Keys.DELETE)
    input_1.send_keys('2019-01-01')
    input_2 = dates_input_fields[1].find_element_by_tag_name('input')
    input_2.send_keys(Keys.CONTROL + "a")
    input_2.send_keys(Keys.DELETE)
    input_2.send_keys('2019-04-15')
    button_element = [q for q in driver.find_elements_by_class_name('d-button-normal') if q.text == "GO"][0]
    button_element.click()
    container = driver.find_element_by_class_name('contentContainer')
    output = []
    for tr in container.find_elements_by_class_name('flex_tr'):
        fields = tr.find_elements_by_xpath(".//div[contains(@class,'data-table-cell')]")
        res = [f.text for f in fields]
        output.append(res)
    return output
```

```python
def scrap_ticker_url(driver, url):
    driver.get(url)
    print (url)
    if 'nyse' in url:
        tickers_info = scrap_nyse_web(driver,url)
    else:
        return
    return pd.DataFrame(tickers_info, columns=['date','open', 'high','low','close','volumne'])
```

```python
final_output = pd.DataFrame(columns=['ticker','date','open','high','low','close','volume'])
for row in df.itertuples():
    try:
        print (row.ticker)
        print (row.url)
        if 'nyse' in row.url:
            df = scrap_ticker_url(driver,row.url) 
        else:
            continue
        df['ticker'] = row.ticker
        final_output = final_output.append(df, sort=True)
    except IndexError:
        print ("Index Error")
    except :
        print ("Error")
```

```python
final_output[['date','open','close','low','high','volumne']].rename(columns={'volumne':'volume'}).to_csv('../output/prices_till_18Apr.csv', index=False)
```
