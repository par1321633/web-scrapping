```python
import requests
```

```python
wikiUrl = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
```

```python
wikiRes = requests.get(wikiUrl)
```

### Using Request Module
Quite lengthy and brute force method. 

```python
data = {"Company":[]}
```

```python
wikiFirstParse = wikiRes.text.split("0001555280")[0]
wikiDataTable = wikiFirstParse.split("component stocks")[3]
hyperLinkSplitWiki = wikiDataTable.split("href=")
```

```python
hyperLinkSplitWiki
```

```python
start = 4
for position in range(len(hyperLinkSplitWiki)):
    if position > start:
        if "nyse" in hyperLinkSplitWiki[position]:
            if "quote" in hyperLinkSplitWiki[position]:
                tempData = hyperLinkSplitWiki[position].split('">')[1].split("</")[0]
                data["Company"].append(tempData)
        elif "nasdaq" in hyperLinkSplitWiki[position]:
            if "symbol" in hyperLinkSplitWiki[position]:
                tempData = hyperLinkSplitWiki[position].split('">')[1].split("</")[0]
                data["Company"].append(tempData)
```

### Using BeautifulSoup Module 


```python
import requests 
from bs4 import BeautifulSoup 
```

```python
soup = BeautifulSoup(wikiRes.content, 'html5lib') 
```

```python
required_div = soup.find('div', attrs={'class' : 'mw-parser-output'})
```

```python
table = required_div.find('table', attrs = {'id': 'constituents'})
```

```python
output = []
for row in table.find_all('tr')[1:]:
    req_fields = row.find_all('td')
    ticker = req_fields[0].text
    url = req_fields[0].a['href']
    company = req_fields[1].text
    industry = req_fields[3].text
    sub_industry = req_fields[4].text
    headquarter = req_fields[5].text
    output.append([ticker.rstrip(), company, industry, sub_industry, headquarter,url])
```

```python
import pandas as pd
```

```python
pd.DataFrame(output, columns=['ticker', 'company', 'industry', 'sub_industry', 'headquarter','url']).to_csv('..\\output\\tickers_sp_500_info.csv')
```
